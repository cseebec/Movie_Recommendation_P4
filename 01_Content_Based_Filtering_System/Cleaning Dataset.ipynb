```python
# Import Dependencies
import pandas as pd
import numpy as np
```

## Cleaning the Movies Dataset

```python
# Read in movie dataframe from Kaggle 
# Link to dataset is: https://www.kaggle.com/tmdb/tmdb-movie-metadata?select=tmdb_5000_movies.csv
movie_file = '01_Starting_Datasets/tmdb_5000_movies.csv'
movies_df = pd.read_csv(movie_file)
movies_df.head()
```

```python
# Get a list of the columns in the dataframe
movies_df.columns
```

```python
# Get a count of how many columns in dataframe
len(movies_df.columns)
```

```python
# Delete columns that will not be used in this code
movies_df = movies_df.drop('budget', axis=1)
movies_df = movies_df.drop('homepage', axis=1)
movies_df = movies_df.drop('original_title', axis=1)
movies_df = movies_df.drop('overview', axis=1)
movies_df = movies_df.drop('popularity', axis=1)
movies_df = movies_df.drop('production_countries', axis=1)
movies_df = movies_df.drop('release_date', axis=1)
movies_df = movies_df.drop('revenue', axis=1)
movies_df = movies_df.drop('runtime', axis=1)
movies_df = movies_df.drop('spoken_languages', axis=1)
movies_df = movies_df.drop('status', axis=1)
movies_df = movies_df.drop('tagline', axis=1)
movies_df = movies_df.drop('vote_average', axis=1)
movies_df = movies_df.drop('vote_count', axis=1)
movies_df.head()
```

```python
# Clean up the genres and keywords columns
for x in range(len(movies_df)):
    # Genre Clean up
    html_genre = movies_df['genres'][x].split(",")
    
    genre_trimmed = ""
    genre = ""
    for i in range(1,len(html_genre),2):
        genre = html_genre[i].replace('"name": "',"")
        genre = genre.replace('"}',"")
        genre = genre.replace(']',"")
        genre_trimmed = genre_trimmed + " " + genre
    
    movies_df['genres'][x] = genre_trimmed
    
    
    # Keyword Cleanup
    html_keyword = movies_df['keywords'][x].split(",")
    
    keyword_trimmed = ""
    keyword = ""
    for i in range(1,len(html_keyword),2):
        keyword = html_keyword[i].replace('"name": "',"")
        keyword = keyword.replace('"}',"")
        keyword = keyword.replace(']',"")
        keyword_trimmed = keyword_trimmed + " " + keyword
    
    movies_df['keywords'][x] = keyword_trimmed
```

```python
# Cleanup the production_companies column
for x in range(len(movies_df)):
    html_comp = movies_df['production_companies'][x].split(",")
    
    comp_trimmed = ""
    comp = ""
    for i in range(0,len(html_comp),2):
        comp = html_comp[i].replace('{"name": "',"")
        comp = comp.replace("[" , "")
        comp = comp.replace('"', '')
        comp_trimmed = comp_trimmed + " " + comp
    
    movies_df['production_companies'][x] = comp_trimmed
movies_df.head()
```

## Cleaning the Credits Dataset

```python
# Read in credits dataframe from Kaggle
# This comes from the same kaggle link as the movie dataframe
credits_file = '01_Starting_Datasets/tmdb_5000_credits.csv'
credits_df = pd.read_csv(credits_file)
credits_df.head()
```

```python
# Delete columns that will not be used in this code
credits_df = credits_df.drop('title', axis=1)
```

```python
# Change movie_id to id
credits_df = credits_df.rename(columns={'movie_id': 'id'})
```

```python
# Cleanup the production_companies column
for x in range(len(movies_df)):
    # Some production companies are a float datatype so I have to skip these
    delete_rows = []
    if type(credits_df['cast'][x]) is float:
        delete_rows.append(x)
    else:
        html_cast = credits_df['cast'][x].split(",")
        
        cast_trimmed =""
        cast = ""
        # Some movies have large casts of more than 50 actors/actresses. 
        # I do not want all these people in the dataframe
        # I only want the stars so I am only going to keep the 5 starring actors/actresses for each movie
        # On the flip side some movies have less than 5 actors/actresses listed
        if len(html_cast)< 40:
            for i in range(5,len(html_cast),7):
                cast = html_cast[i].replace('"name": ','')
                cast = cast.replace('"','')
                cast_trimmed = cast_trimmed + " " + cast
        else:
            for i in range(5,40,7):
                cast = html_cast[i].replace('"name": ','')
                cast = cast.replace('"','')
                cast_trimmed = cast_trimmed + " " + cast
    
        credits_df['cast'][x] = cast_trimmed

credits_df.head()
```

## Merge the Datasets

```python
# Convert id in movies_df to string 
movies_df['id'] = movies_df['id'].apply(str)
```

```python
# Merge the dataframes
merge_df = pd.merge(movies_df, credits_df, on="id", how="inner")
print(len(merge_df))
```

```python
merge_df.count()
```

Since all columns have the same count I know there are no null values

```python
# Make the index its own column
merge_df["index"] = merge_df.index
```
